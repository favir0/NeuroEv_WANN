```python
import pickle
from population import Population
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

file_name_to_load = "outputs/PopulationDumps/cancer_evo_pop"

def load(filename: str) -> Population:
    with open(filename, 'rb') as f:
        population = pickle.load(f)
    return population

population = load(file_name_to_load)

# 1.    
champion_fitness = [genome.fitness for genome in population.champions]

# 2.    
average_fitness = population.avg_fitness

# 3.  ,         
min_connections = population.min_connections
max_connections = population.max_connections
champion_connections = [len(genome.connections) for genome in population.champions]

# 4.  3   
min_nodes = population.min_nodes
max_nodes = population.max_nodes
champion_nodes = [len(genome.nodes) for genome in population.champions]

# 5.   ,     
max_species_age = population.max_species_age
avg_species_age = population.avg_species_age

# 7.    
species_amount = population.species_amount

# 8.    
compatibility_thresholds = population.compatibility_thresholds

#  
plt.figure(figsize=(20, 15))

# 1.    
plt.subplot(4, 2, 1)
plt.plot(champion_fitness, label='Champion Fitness')
plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.title('Champion Fitness per Generation')
plt.legend()

# 2.    
plt.subplot(4, 2, 2)
plt.plot(average_fitness, label='Average Fitness')
plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.title('Average Fitness per Generation')
plt.legend()

# 3.  ,        
plt.subplot(4, 2, 3)
plt.plot(min_connections, label='Min Connections')
plt.plot(max_connections, label='Max Connections')
plt.plot(champion_connections, label='Champion Connections')
plt.xlabel('Generation')
plt.ylabel('Connections')
plt.title('Connections per Generation')
plt.legend()

# 4.     
plt.subplot(4, 2, 4)
plt.plot(min_nodes[1:], label='Min Nodes')
plt.plot(max_nodes[1:], label='Max Nodes')
plt.plot(champion_nodes[1:], label='Champion Nodes')
plt.xlabel('Generation')
plt.ylabel('Nodes')
plt.title('Nodes per Generation')
plt.legend()

# 5.   ,     
plt.subplot(4, 2, 5)
plt.plot(max_species_age, label='Max Species Age')
plt.plot(avg_species_age, label='Average Species Age')
plt.xlabel('Generation')
plt.ylabel('Age')
plt.title('Species Age per Generation')
plt.legend()

# 6.       
plt.subplot(4, 2, 6)
color = 'tab:blue'
plt.plot(species_amount, label='Number of Species', color=color)
plt.xlabel('Generation')
plt.ylabel('Number of Species')
plt.title('Number of Species and Compatibility Threshold per Generation')
plt.legend(loc='upper left')

ax2 = plt.gca().twinx()
color = 'tab:red'
ax2.plot(compatibility_thresholds, label='Compatibility Threshold', color=color)
ax2.set_ylabel('Compatibility Threshold')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```
